## Loop through

```python
filedirectory = ["raw_data", "combined_data"]
filename = "sat_data_combined"
df = pd.read_excel(f"data/survey_data/{filedirectory[1]}/{filename}.xlsx", index_col=0)
df.reset_index(inplace=True)
#df.drop(columns=['Optional', 'Timestamp'], inplace=True)
df.drop(columns=['index'], inplace=True)
```

```python
df.columns
```

```python
def loop_data(df, file_name, loop_type="bm", row_n = 1, num_of_sentences=1, save_data=True, preferred_save_type="csv"):
    if loop_type.lower() == "bm":
        src = "en"
        tgt = "de"
        bm = paraphrase.BacktranslationMachine(src=src, tgt=tgt)
        columns = df.columns.values
        input_sentences = df.iloc[row_n].values
        generated_sentence = bm.backtranslation(input_sentences)
        df2 = pd.DataFrame(generated_sentence).T
        df2.columns = list(df.columns.values)
        #frames = [df, df2]
        #df2 = pd.concat(frames)

    else:
        pm = paraphrase.ParaphraseMachine()
        columns = df.columns.values
        input_sentences = df.iloc[row_n].values
        print(f"Paraphrase begin for row {row_n}")
        generated_sentence = pm.paraphrase(list(input_sentences), num_of_sentences=num_of_sentences)
        print("Paraphrase done")
        if num_of_sentences == 1:
            df2 = pd.DataFrame(generated_sentence).T
            df2.columns = list(df.columns.values)
        else:
            generated_nested_sentence = [generated_sentence[i:i+num_of_sentences] for i in range(0, len(generated_sentence), num_of_sentences)]
            generated_dict = {}
            for i, inner_list in enumerate(generated_nested_sentence):
                generated_dict[columns[i]] = inner_list
            df2 = pd.DataFrame(generated_dict)
        
        
    if save_data:
        if preferred_save_type == "csv":
            print("Saving as csv.")
            df2.to_csv(f"data/survey_data/processed_data/{filename}_{loop_type}_row{row_n}.csv", mode='a', header=False, index=False, encoding="ISO-8859-1")
        else:
            print("Saving as excel.")
            df2.to_excel(f"data/survey_data/processed_data/{filename}_{loop_type}_row{row_n}.xlsx", index=False)
        #df2.to_csv(f"sat_{loop_type}_paraphrased_data.csv")
    print(f"Done paraphrasing data with {loop_type} machine.")
    return df2
        
```
